```python
# import libraries
import pandas as pd
from scipy.stats import uniform
from scipy.stats import randint
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from scipy import stats

class LinearRegression(linear_model.LinearRegression):
    """
    LinearRegression class after sklearn's, but calculate t-statistics
    and p-values for model coefficients (betas).
    Additional attributes available after .fit()
    are `t` and `p` which are of the shape (y.shape[1], X.shape[1])
    which is (n_features, n_coefs)
    This class sets the intercept to 0 by default, since usually we include it
    in X.
    """

    def __init__(self, *args, **kwargs):
        if not "fit_intercept" in kwargs:
            kwargs['fit_intercept'] = True
        super(LinearRegression, self)\
                .__init__(*args, **kwargs)

    def fit(self, X, y, n_jobs=1):
        self = super(LinearRegression, self).fit(X, y, n_jobs)

        sse = np.sum((self.predict(X) - y) ** 2, axis=0) / float(X.shape[0] - X.shape[1])
        se = np.array([
            np.sqrt(np.diagonal(sse[i] * np.linalg.inv(np.dot(X.T, X))))
                                                    for i in range(sse.shape[0])
                    ])

        self.t = self.coef_ / se
        self.p = 2 * (1 - stats.t.cdf(np.abs(self.t), y.shape[0] - X.shape[1]))
        return self

def plotManhattan(genes,pvalues,chromosomes,title="test plot",savepath="manhattan_test.png"):
    """
    Generate and save manhattan plot
    
    genes: list of genes indexes
    pvalues: list of the p-values corresponding to the genes
    chromosomes: list of the chromosomes corresponding to the genes
    title: title of the plot
    savepath: path where the plot is saved
    """
    df = pd.DataFrame(
        {
            'gene' : genes,
            'pvalue' : pvalues,
            'chromosome' : chromosomes
        })
    
    # -log_10(pvalue)
    nbchrom = len(df.chromosome.unique())
    df['minuslog10pvalue'] = -np.log10(df.pvalue)
    df.chromosome = df.chromosome.astype('category')
    df.chromosome = df.chromosome.cat.set_categories(['ch-%i' % i for i in range(1,nbchrom+1)], ordered=True)
    df = df.sort_values('chromosome')

    # How to plot gene vs. -log10(pvalue) and colour it by chromosome?
    df['ind'] = range(len(df))
    df_grouped = df.groupby(('chromosome'))

    # manhattan plot
    fig = plt.figure(figsize=(14, 8)) # Set the figure size
    
    
    ax = fig.add_subplot(111)
    colors = ['gold','deepskyblue','hotpink', 'limegreen']
    x_labels = []
    x_labels_pos = []
    for num, (name, group) in enumerate(df_grouped):
        group.plot(kind='scatter', x='ind', y='minuslog10pvalue',color=colors[num % len(colors)], ax=ax)
        x_labels.append(name)
        x_labels_pos.append((group['ind'].iloc[-1] - (group['ind'].iloc[-1] - group['ind'].iloc[0])/2))
    ax.set_xticks(x_labels_pos)
    ax.set_xticklabels(x_labels)

    # set axis limits
    bonfcorrLimit = -np.log10(0.05/df.shape[0])
    print(bonfcorrLimit)
    ax.set_xlim([0, len(df)])
    #ax.set_ylim([0, max(df["minuslog10pvalue"].max(),bonfcorrLimit+0.5)])
    ax.set_ylim([0, 30])

    # x axis label
    ax.set_xlabel('Chromosome')
    
    # Significant genes separation with bonferroni correction
    plt.axhline(y = bonfcorrLimit, color = 'r', linestyle = '-', label='separation with Bonferroni correction')

    # show the graph
    plt.title(title, y=-0.2)
    plt.legend()
    plt.savefig('manhattan_test.png')
    plt.show()

def getpvalue(traitValues,allelePairValues,plot=False):
    """
    traitValues: list of the values of the corresponding trait
    allelePairValues: list of 0,1 or 2 corresponding to the nature of the allele pair
    plot: if we want to plot the result
    """
    reg = LinearRegression().fit(allelePairValues, traitValues)
    if plot:
        plt.scatter(allelePairValues, traitValues,color='g')
        plt.plot(allelePairValues, reg.predict(allelePairValues),color='k')
        plt.show()
    return reg.p[0][0]

def getpvalueformanhattan(Xmat,trait,row=0,plot=False):
    df = pd.DataFrame(
            {
                'trait' : trait,
                'allelePair' : Xmat.loc[row]
            })
    df.dropna(inplace=True)
    df['allelePair2'] = [[i] for i in df["allelePair"]]

    X_train = np.array(df["allelePair2"].values.tolist())
    y = np.array(df["trait"].values.tolist()).reshape(-1, 1)
    return getpvalue(y,X_train,plot=plot)

def getpvalues(Xmat,y):
    """
    Xmat: matrix of allele pairs for each gene (rows) for each accession (columns)
    y: vector of the trait values
    """
    pvalues = []
    
    for i in range(Xmat.shape[0]):
        if not i%500:
            print(i)
        pvalues.append(getpvalueformanhattan(Xmat,y,i))
    return pvalues
```

```python
pheno = pd.read_csv("../../Ressources/pheno.csv")
pheno = pheno[["HybID","NSFTVID","Seed.number.per.panicle","Amylose.content","Straighthead.suseptability"]]
pheno
```

```python
geno = pd.read_csv("../../Ressources/geno.csv")
genoMat = geno[[str(i) for i in range(1,414)]]
genoMat.shape[0]
```

```python
genes = geno["marker"]
chromosomes = pd.Series([f"ch-{i}" for i in geno["chrom"]])
pvalues = getpvalues(genoMat,list(pheno["Seed.number.per.panicle"]))

df = pd.DataFrame(
        {
            'gene' : genes,
            'pvalue' : pvalues,
            'chromosome' : chromosomes
        })
df.to_csv("pvaluesdataframe.csv")
```

```python
plotManhattan(genes,pvalues,chromosomes)
```
